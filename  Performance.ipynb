```python
import numpy as np
import pandas as pd
from KF import KFsim
import matplotlib.pyplot as plt
import seaborn as sns
from utils.PQ_matrices import P_cov
```

```python
Amax = 4*32
Pmax = 0.1
P_o  = 0.5
σ_m2 = (((Amax**2)/3)*(1+4*Pmax - P_o))
```

```python
T = 0.001
α = 10

sigma_ratio =  10e-6
σ_r2        =  σ_m2/sigma_ratio
σ_r         =  σ_r2**0.5
```

```python
allps =[]
Ts = []
```

```python
while T<101:
    Ps, cov = KFsim(σ_m2**0.5,σ_r,α,T,Amax)
    allps.append(Ps)
    T= T * 1.10
    Ts.append(T)
    
```

```python
p11 = []
for T in range(len(allps)):
    p11.append(np.mean(allps[T][0][0][0]))#) for i in range(600)])/(σ_r2))
    
p11df = pd.DataFrame(p11)
p11df["ts"] = Ts
p11df.columns = ["ps", "ts"]
sns.lineplot(x = p11df.ts, y = p11df.ps)
plt.xscale('log')
plt.yscale('log')
```
